## 4) Trapezoidal rule

### a)

```python
a=0
b=2
x = np.linspace(a,b,10)
def f(x):
    return 7*x**5 +x**3 +1

#f_int=function[0]
def trapeze (f,x):
    nx = x.shape[0]
    delta = (x[-1]-x[0])/(nx-1)
    f_int = 0
    for i in range(1,nx-1):
        f_int += f(x[i])
    f_int+= (f(x[0]) + f(x[-1]))/2
    return f_int*delta
f_integrate = trapeze (f,x)
print(f_integrate)
```

```python
def trapeze_the_best(f,x): 
    nx = x.shape[0]
    delta = (x[-1]-x[0])/(nx-1)
    F_the_best = 0
    F_the_best += (f(x[0]) + f(x[-1]))/2
    F_the_best +=  np.sum(f(x[1:-1]))
    return(F_the_best*delta)
f_best = trapeze_the_best(f,x)
print(f_best)
```

## 6) Dichotomy

```python
a=0
b=2
x = np.linspace(a,b,10)

def f(x):
    """ define the function to sudy """
    return(x-np.exp(-x))

def dichotomy(f,eps):
    #intervalle sur lequel je veux travailler
    intervall = np.abs(x[-1] - x[0])
    if f(x[0])*f(x[-1])<0:
        
        while (x[-1]-x[0])>eps:
            mid_pt = (x[0]+x[-1])/2
            
            if f(mid_pt)*f(x[0])<=0:
                x[-1] = mid_pt 
                intervall = np.abs(x[-1] - x[0])

            else:
                x[0] = mid_pt 
                intervall = np.abs(x[-1]-x[0])
    else : 
        print('choose another intervall, no solutions in this one')
    return(0.5*(x[0]+x[-1]))
           
sol = dichotomy(f,0.0005)
print(sol)
```

## 7) Sorting algorithm

```python
import random

def TriaBulles(L):
    Ech = True
    while Ech:
        Ech = False
        for i in range(len(L)-1):
            if L[i] > L[i+1]:
                L[i],L[i+1] = L[i+1],L[i]
                Ech = True

    return(L)

L = [random.randrange(0,100) for i in range(10)]
print(L)
L_trie = TriaBulles(L)
print(L_trie)
```
