```python
import numpy as np
import matplotlib.pyplot as plt
import random

#Set plot style
plt.rcParams.update({
    "lines.linewidth": 2,
    "axes.grid": True})

n_pts = 25

#generate coordinates of random points 
x = np.random.random_sample(n_pts)*10
y = np.random.random_sample(n_pts)*10

# We attribute an integer label from 0 to 9 for each point
# corresponding to its place in the x and y arrays.
# for instance if the lowest value is at the 2nd position and the 2nd lowest value at 9th position, 
#it will goes like : [2,9,...] 
# Then we will work on sorted arrays along x and y axis.
xi_sorted = np.argsort(x)
yi_sorted = np.argsort(y)
#print(xi_sorted)

# These arrays help us to find where an
# point is located in the sorted array.

whr_xi = np.argsort(xi_sorted)
whr_yi = np.argsort(yi_sorted)
#print(whr_xi,whr_yi)
plt.scatter(x,y, c= np.arange(n_pts), marker = '*', 
            cmap = 'plasma', 
            label='original points')
plt.scatter(whr_xi,whr_yi, c= np.arange(n_pts), marker= '^', cmap = 'plasma',label='sorted points')
plt.scatter (whr_xi[0], whr_yi[0], s=200,facecolors = 'none', edgecolors='r')
plt.legend(loc =4 )
plt.show()

#we create a matrix that contains the points sorted
#so xi_srt are the colums and yi_srt are the lines 
#this matrix is going to be quite empty, but it's easier to select the points in moving in a matrix
#we will select the nearest points (left and right) from a random point and define a perimeter
#so we will reduce the field of research more and more 

Matrix = np.zeros((n_pts,n_pts))
row = 0
colums = 0



# we look at the points sorted except those that are at the edges
for i in range (1, n_pts-1):
    #we take one random point from the sorted ones
    x_ini = xi_sorted[i]
    #we look what are the nearest points of it, meaning that we look what term is at his left,
    #and what term is a his right
    left_x_ini = np.abs(xi_sorted[i-1] - x_ini)
    right_x_ini = np.abs(xi_sorted[i+1] - x_ini)
    
    #now I have to find the corresponding in y-axis
    #I know for the x-axis the left and rights points
    
    left_y_ini = whr_yi[left_x_ini]
    right_y_ini = whr_yi[right_x_ini]
    print('sorted points for x-axis',left_x_ini,right_x_ini)
    print('sorted points for y-axis',left_y_ini,right_y_ini)
```
