````python
def Disque(x,y,R) :
    r=((x**2)+(y**2))**.5
    return np.sign(R-r)

x=np.linspace(-5,5,1000)
y=np.linspace(-5,5,1000)
X,Y=np.meshgrid(x,y)
    
f = Disque(X,Y,2)
    
pyplot.pcolormesh(X,Y,f,cmap="coolwarm")
pyplot.show()
````

````python
def Disque_new(x,y,R,a,b) :
    r=(((x-a)**2)+((y-b)**2))**.5
    return np.sign(R-r)

x=np.linspace(-5,5,1000)
y=np.linspace(-5,5,1000)
X,Y=np.meshgrid(x,y)
    
f = Disque_new(X,Y,2,2,3) + Disque_new(X,Y,2,-2,-3)
    
pyplot.pcolormesh(X,Y,f,cmap="coolwarm")
pyplot.show()
````

````python
def Rectangle(x,y,a,b):
    Z=np.ones(x.shape)
    Z[np.abs(x)> (a/2)] = 0
    Z[np.abs(y)>(b/2)] = 0
    return Z
    
x=np.linspace(-5,5,1000)
y=np.linspace(-5,5,1000)
X,Y=np.meshgrid(x,y)

f=Rectangle(X,Y,1,3)

pyplot.pcolormesh(X,Y,f,cmap="coolwarm")
pyplot.show()
````

## Exercice 6: 
Un petit bouchon $S_1$ est placé à la surface d'une étendue d'eau initialement au repos. Il est animé d'un mouvement vertical sinusoidal de fréquence $f$. La perturbation de la surface de l'eau induite par le bouchon est également sinusoidale et se propage loin de $S_1$ avec une vitesse $c$: c'est une onde. Soit $z$ la variation de hauteur de l'eau.
\begin{equation}
z(x,y,t)=z_0 cos(2 \pi f (t-S_1M/c))
\end{equation}  
où $z_0$ est l'amplitude de l'onde, $f$ la fréquence, $t$ l'instant d'observation, c la vitesse de propagation de l'onde et $S_1M$ la distance de la source $S_1$ (le bouchon) au point $M(x,y)$ situé sur l'étendue d'eau. 
\begin{equation}
S_1M=\sqrt{(x_1-x)^2+(y-y_1)^2}
\end{equation}
 

1. Créer une fonction calculant la variation de hauteur $z$ de l'eau en fonction des coordonnées x,y et des autres données du problème. $z_0$, une fréquence $f$ et un instant $t$.
2. Représenter graphiquement $z$ à un instant $t$ donné ($S_1$ est placée en $x_1=0$ et $y_1=0$). Remarque: $\lambda=c/f$ est la longueur d'onde. Il faut que les échelles en $x$ et $y$ permettent de voir quelques longueurs d'onde. On prendra par exemple $c=1 m.s^{-1}$ et $f=5 Hz$    
3. Que se passe-t-il lorsque l'on change l'instant $t$ d'observation ?
4. Enregistrer les valeurs de $z$ obtenue à $t=1s$ dans un fichier `champ1.npy`.
5. Reprendre les questions avec une source $S_2$ placée en $x_1=0$ et $y_1=2 m$. On enregistrera les résultats dans un fichier `champ2.npy`.
6. Ecrire un programme qui récupère les données de ces deux fichiers et effectue la somme des perturbations de hauteur produites par les deux sources. Afficher le résultat. 

````python
import numpy as np
from matplotlib import pyplot

def z(x, y, t, z0, f, c, x1, y1):
    S1M = (((x1-x)**2) + (y-y1)**2)**.5
    return z0 * np.cos(2*np.pi*f*(t-S1M/c))

x=np.linspace(-4,4,1000)
y=np.linspace(-4,4,1000)
X,Y = np.meshgrid(x,y)

Z=z(X, Y, 1, 1, 5, 1, 0, 0)

pyplot.pcolormesh(X,Y,Z)
pyplot.show()

np.save("champ1.npy",Z)
````
