```python
import os
import numpy as np
import datetime
from matplotlib import pyplot as plt
from mod1 import orbitalparam,orbital_timeseries,ephemeride,planet_average,insolation,plot_y,plot_yt

# (a) insolation for present-day conditions + LOD
kyear=0
ecc, epsilon, omega = orbitalparam(kyear)
resu0 = insolation(kyear,1) # CTRL
lon=resu0['lon'];lat=resu0['lat']
solar0_ave=planet_average(resu0['solar_am'],lat,lon,lat)
print ("annual global mean insolation =", '%.4f' %solar0_ave ," W/m2")
```

```python
# (b) turn perihelion 180 deg
resu1 = insolation([0.017236,23.446,281.37+180],1)
plot_yt(resu1['solar']-resu0['solar'],'Mean daily insolation difference W/m2')
plot_y(resu1['solar'],resu0['solar'],'present day','perihelion turned 180 deg')
solar1_ave = planet_average(resu1['solar_am'], lat, lon, lat)
#changing the perihelion
"""resu2 = insolation([0.017236,23.446,281.37+190],1)
plot_yt(resu2['solar']-resu0['solar'],'Mean daily insolation difference W/m2')
plot_y(resu2['solar'],resu0['solar'],'present day','perihelion turned 180 deg')
solar2_ave = planet_average(resu2['solar_am'], lat, lon, lat)
print ("annual global mean insolation=", '%.4f' %solar2_ave ," W/m2")"""
print ("annual global mean insolation=", '%.4f' %solar1_ave ," W/m2")
```

```python
# (c) obliquity 26 deg
#changing the obliquity means changing the amplitude of the seasons, see graph untitled Annual mean incoming solar radiation
#rising obliquity is rising temp at poles
#obliquity is the inclination of a planet from the ellipse 
resu2 = insolation([0.017236,26,281.37],1)
plot_yt(resu2['solar']-resu0['solar'],'Mean daily insolation difference W/m2')
plot_y(resu0['solar'],resu2['solar'],'present day','obliquity 26 deg')
solar2_ave = planet_average(resu2['solar_am'], lat, lon, lat)
print ("annual global mean insolation=", '%.4f' %solar2_ave ," W/m2")
```

```python
# (d) obliquity 60 deg
# here insolation becomes greater at poles than at the equator
resu3 = insolation([0.017236,60,281.37],1)
plot_yt(resu3['solar']-resu0['solar'],'Mean daily insolation difference W/m2')
plot_y(resu0['solar'],resu3['solar'],'present day','obliquity 60 deg')
solar3_ave = planet_average(resu3['solar_am'], lat, lon, lat)
print ("annual global mean insolation=", '%.4f' %solar3_ave ," W/m2")
```

```python
# (e) Last glacial Inception
#115 thouthands years B JC, glaciar period
#there were less insolation at 115*10**3 than today 
#to define a glacial inception, we need to have several cold periods
orbital_timeseries(np.arange(0, 501, 1), 65)
resu4=insolation(115,ifig=1);
plot_yt(resu4['solar']-resu0['solar'],'Mean daily insolation difference W/m2')
plot_y(resu0['solar'],resu4['solar'],'present day','Inception of the last glacial period')
solar4_ave = planet_average(resu4['solar_am'], lat, lon, lat)
print ("annual global mean insolation=", '%.4f' %solar4_ave ," W/m2")
```

```python
# just for fun: sunrise and sunset at Brest
mon=datetime.datetime.now().month
day=datetime.datetime.now().day
year=datetime.datetime.now().year
ephemeride(day,mon,year,-4.4833,48.4,2,True)
ephemeride(day,mon,year,-4.4833,48.4,2,False)
```

```python
# just for fun: length of the day in Brest for the different situations seen above
plt.figure(figsize=(10,5))
plt.plot(resu0['days'],resu0['LOD_Brest'],'k',label='present day')
plt.plot(resu0['days'],resu1['LOD_Brest'],'r',label='perihelion turned 180 deg')
plt.plot(resu0['days'],resu2['LOD_Brest'],'g',label='obliquity 26 deg')
plt.plot(resu0['days'],resu3['LOD_Brest'],'b',label='obliquity 60 deg')
plt.plot(resu0['days'],resu4['LOD_Brest'],'skyblue',label='115 kyr ago')
plt.legend(loc='best',fontsize=12)
plt.xlabel('days',fontsize=14);plt.ylabel('hours',fontsize=14)
plt.title('Length of the day in Brest',fontsize=14)
plt.xticks(resu0['days_mid'],('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'))
plt.show();
```
