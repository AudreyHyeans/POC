# ACTIVITY 2: Geostrophy, thermal wind, eddy momentum fluxes and the maintenance of surface westerlies

The objective of this exercise is 1/ to use the thermal wind balance
to determine how is the large-scale atmospheric circulation close to
geostrophy and 2/ to explore the crucial role of transient eddy

  Atmospheric Research) reanalysis that provides a globally gridded

winds (we will focus on the zonal component). The corresponding netcdf
files are available in the folder data I

I have also computed the momentum transport from daily mean zonal and

```python
# 3/ Compute the layer-mean atmospheric temperature
#    int T d lnp / int d lnp
# We will call this variable tbar.
# There are nz-1 layers. Hence the dimension of tbar is (nz-1,ny,nx)

tvar=resu_t_am['T'] # annual mean temperature at "pres" levels (nz levels)
print(np.shape(tvar))
#on redimensionne le tableau dlnp avec nx,ny,nz.Puis on veut nz,nx,ny donc on crit (2,0,1)
dlnp = np.transpose(np.tile(np.diff(np.log(pres)),(73,144,1)),(2,0,1))
tbar = np.zeros((nz-1,ny,nx))
tmid=0.5*(tvar[:-1,:,:] + tvar[1:,:,:])        
"""    
#il faut trouver les dp et on a Tmid = 0.5*(T^k+T^k+1)
T_mid = 0
T_mean = 0
pres_ln = np.log(pres)

dp = np.diff(pres_ln)
      
T_mid = 0.5*(tvar[:,0,0] + tvar[:+1,0,0])
T_mean = np.cumsum(T_mid)

R=287
g0 = 9.81
z = -(R*T_mean)/g0 * np.sum(dp)

print(np.shape(z))
"""
```

```python
# 4/ Compute the geopotential height using the hypsometric
rd=287.0
g0=9.806
zerodk=273.15
z_hypso = np.zeros((nz,ny,nx))
z_hypso[0,:,:] = resu_z_am['Z'][0,:,:]
pres_redim = np.transpose(np.tile(pres,(ny,nx,1)),(2,0,1))
z_hypso[1:,:,:] = z_hypso[0,:,:] + (rd*(tbar+zerodk)/g0)*np.log(pres_redim[0,:,:]/pres_redim[1:,:,:])
```

```python
# Compare the zonal mean geopotential height in observations to that you just obtained from
# the hypsometric equation at 250 mb
k0=8 # k0=8 corresponds to 250mb
plt.figure(figsize=(8,4))
plt.plot(lat,np.mean(resu_z_am['Z'][k0,:,:],axis=1),'k')
plt.plot(lat,np.mean(z_hypso[k0,:,:],axis=1),'r')
plt.xlabel('Latitude',fontsize=14)
plt.ylabel('m',fontsize=14)
tcar=str('Zonal and annual mean geopotential height at p = '+str(pres[k0])+' hPa')
plt.title(tcar,fontsize=14)
plt.show(block=False)
```

```python
# 3/ Compute geostrophic wind from geopotential height
#    f k x u = -g grad Z
#    Note that a warning will be produced for f=0 (equator)

# Let us call ug the geostrophic wind
phi = np.deg2rad(lat)
ug=np.zeros((nz,ny,nx))
omega = 2*np.pi/86400
f = 2*omega*np.sin(phi)
a=6.37e+06
#grad(p)
dlat = np.diff(lat)

np.shape(lat)
 
#np.shape(pres)
for i in range (1,ny-1,1):
    ug[:,i,:] = -(g0/(a*f[i])) * (resu_z_am['Z'][:,i,:] - resu_z_am[:,i-1,:])/(phi[i+1]-phi[i-1])

```

```python
# plot zonal average geostrophic winds
# and compare to observations
plt.figure(figsize=(10,12))
plt.subplots_adjust(hspace=0.3)
plt.subplot(3,1,1)
varn=ug
c=plt.contourf(lat,pres,np.mean(varn,axis=2),31,cmap=plt.cm.jet)
plt.colorbar(c)
plt.clim(-15,40)
plt.gca().invert_yaxis()
plt.ylabel('pressure (mb)',fontsize=14)
plt.title('Computed geostrophic Wind (m/s)')
plt.subplot(3,1,2)
varn=resu_u_am['U']
c=plt.contourf(lat,pres,np.mean(varn,axis=2),31,cmap=plt.cm.jet)
plt.colorbar(c)
plt.clim(-15,40)
plt.gca().invert_yaxis()
plt.ylabel('pressure (mb)',fontsize=14)
plt.title('Observed Wind (m/s)')
plt.subplot(3,1,3)
varn=ug-resu_u_am['U']
c=plt.contourf(lat,pres,np.mean(varn,axis=2),31,cmap=plt.cm.jet)
plt.colorbar(c)
plt.gca().invert_yaxis()
plt.xlabel('Latitude',fontsize=14)
plt.ylabel('pressure (mb)',fontsize=14)
plt.title('Difference between Ug and Observed Wind (m/s)')
plt.show(block=False)
```

```python
# Deduce a Rossby number (called Rossby below)
# The Rossby number is a measure of the departure of the flow from geostrophy
# The smaller the Rossby number the closer the circulation to geostrophy

plt.figure()
scale=np.linspace(0,1,100) # we focus on the values between 0 and 1
c=plt.contourf(lat,pres,Rossby,scale,cmap=plt.cm.jet,extend='max')
plt.colorbar(c)
plt.clim(0,scale[-1])
plt.gca().invert_yaxis()
plt.xlabel('Latitude',fontsize=14)
plt.ylabel('pressure (mb)',fontsize=14)
plt.title('Rossby number')
plt.show(block=False)
```

```python
# 4/ compute divergence
# Let us call this divergence "div_eddy_fluxes"
# Note the the dimension in latitude is nlat-1 (because we compute a gradient from a field which has
# nlat points in the latitude)
# The divergence is thus defined at mid-points in latitude --> latitude lat_mid below
div_eddy_fluxes=np.zeros(nlat-1)
lat_mid=0.5*(lat[:-1]+lat[1:])
```

```python
# 5/ make a plot with the convergence (which is minus divergence)
fig=plt.figure(figsize=(8,10))
plt.subplots_adjust(hspace=0.15)
ax1=plt.subplot(2,1,1)
mabs=np.amax([np.abs(np.amin(uv_tra)),np.amax(uv_tra)])
scale=np.linspace(-mabs,mabs,51)
c1=plt.contourf(lat,pres,uv_tra,scale,cmap=plt.cm.seismic)
plt.rcParams['contour.negative_linestyle'] = 'dashed'
plt.contour(lat,pres,uv_tra,21,colors='k')
plt.ylabel('pressure (mb)',fontsize=14)
plt.gca().invert_yaxis()
plt.title(r"Transient Eddy Momentum Fluxes $[\overline{u^\prime v^\prime}]\cos^2\phi$ (m$^2$s$^{-2}$)",fontsize=14)
plt.xlim(-90,90)
ax2=plt.subplot(2,1,2)
plt.plot(lat_mid,-div_eddy_fluxes,'k')
plt.plot([-90,90],[0,0],'k--')
plt.title("Transient Eddy Momentum Fluxes Convergence",fontsize=14)
plt.xlabel('Latitude',fontsize=14)
plt.ylabel('Pa',fontsize=14)
plt.xlim(-90,90)
box1 = ax1.get_position()
box2 = ax2.get_position()
ax1.set_position([0.125,0.55,0.7,0.35])
ax2.set_position([0.125,0.10,0.7,0.35])
box1 = ax1.get_position()
box2 = ax2.get_position()
cbaxes = fig.add_axes([0.01+box1.x1,box1.y0, 0.02, box1.y1-box1.y0]) 
cb1=plt.colorbar(c1,orientation='vertical',cax=cbaxes)
plt.show(block=False)
```

```python
# 6/ Deduce an estimation of the zonal mean surface zonal wind speed
#    assuming a linear drag law, i.e. taux=rhoa*cdrag*utt*us

cdrag=2.0e-3 # typical surface drag coefficent
rhoa=1.25 # surface air density
utt=5.0 # turbulent wind speed
usurf=-div_eddy_fluxes/(rhoa*cdrag*utt)

plt.figure(figsize=(10,6))
plt.plot(lat_mid,usurf,'k')
plt.plot([-90,90],[0,0],'k--')
plt.xlim(-90,90)
plt.ylabel(r'm s$^{-1}$',fontsize=14)
plt.title('Estimated zonal mean surface zonal wind',fontsize=14)
plt.xlabel('Latitude',fontsize=14)
plt.show(block=False)
```
