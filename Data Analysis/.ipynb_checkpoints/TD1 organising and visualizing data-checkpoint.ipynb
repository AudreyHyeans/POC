# 1: Finding outliers in Argo data

You will analyse temperature and salinity data collected by [Argo](https://argo.ucsd.edu/) floats. These are instruments that drift independently in the ocean at a depth of 1000m. Every 10 days, they descend to 2000m, and then rise back up to the surface, collecting measurements of the temperature and salinity of the water column as they do so. Each set of these measurements (from 2000m depth to surface) is called a _profile_. They then return this information by satellite, before re-descending to 1000m and drifting for another 10 days.

![Schematic of an Argo float cycle](https://argo.ucsd.edu/wp-content/uploads/sites/361/2020/06/float_cycle_1.png)

Argo floats have revolutionised observational oceanography since they were first deployed (in 2000) by providing many more measurements than at any previous time in history. However, the floats often have technical problems, and many people work on controlling the quality of these data, to be sure that the measurements returned are accurate.

In this first part, you will look at measurements returned by Argo floats in a given region. You will use data that have not been quality controlled (meaning that errors have not yet been corrected) and identify any Argo floats that might be having technical problems.

### 1.3: The data set contains measurements taken over a six month period. For each month, calculate the monthly mean and standard deviation of the temperature and salinity averages $\bar{T}_{500}$ and $\bar{S}_{500}$ calculated in _\#1.2_ above. Print out these values.

There is a variable called ```TIME``` in the data file. This contains the date in numpy's ```datetime``` format. To extract the information about the month, you can use a built-in function: if you take the first element of the time array, for example, ```ds['TIME'][0].month``` will return the month of this datetime value. You will need to loop through each element of the ```TIME``` array, and use this function to obtain the value of the month and store it in a second array.

To calculate the monthly means: for June, you will select all the measurements taken in June, and report the mean of these measurements and their standard deviation for both temperature and salinity. Then do the same for the other months.

````python
# enter your code here
time = ds['TIME']
month = []
m_Te = []
m_Sa = []
sdt_Te = []
sdt_Sa = []
for i in range(len(time)):
    month.append(time[i].month)
month = np.array(month)
for m in range (6,12):
    m_Te = np.append(m_Te,np.mean(mean_Te[np.where(month==m)]))
    sdt_Te = np.append(sdt_Te, np.std(mean_Te[np.where(month==m)],ddof=1))
    
    m_Sa=  np.append(m_Sa,np.mean(mean_Sa[np.where(month==m)]))
    sdt_Sa = np.append(sdt_Sa,np.std(mean_Sa[np.where(month==m)],ddof=1))
    print(m_Te)
````

# 2: Separating measurements of different water masses

Water masses are large bodies of water that have particular physical or chemical properties that allow us to isolate and identify them. For example, a certain volume of water might have very high salinity (Mediterranean Water), or very low temperature (Antarctic Surface Water) compared to its surroundings. These signature properties are set when the water mass is formed, usually by the atmosphere acting on the ocean in a certain way (causing strong cooling, or evaporation, for example), or by interaction with sea ice. Water masses are widely studied in research; changes in water mass properties can give us insights into the effects of climate change, to give just one example.
