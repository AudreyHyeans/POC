# Introduction

When we make observations in oceanography, or when we run numerical models, we obtain data. If we make good judgements about how to interpret that data, it is more likely that we will correctly interpret what is happening physically. If we make bad judgements about how to interpret the data, we might arrive at the wrong conclusions about the physics. The aim of these data analysis classes is to help you to make good judgements about how to interpret data. To do that, first we need to learn some basic statistics.

## 1: Variables and data

Variables are quantities that do not take a fixed value (e.g. temperature, salinity, depth are all variables; the speed of light and the Stefan-Boltzmann constant are not variables).

- Categorical or qualitative variables have non-numerical values (e.g. water masses: {Subantarctic Mode Water, Mediterranean Water, etc}, seasons: {spring, summer, autumn, winter})
- Quantitative variables have numerical values (e.g. temperature). They can be discrete (the possible values are fixed), or continuous (any value is possible within a given range)
- Data are the values that are taken by a variable

We can make toy data using a random number generator in python. Click within the box ("cell") below, and then press <br><kbd>shift &#8679;</kbd> and <kbd>return &#8629;</kbd> together to run the code

````python
import numpy as np # this library contains a random number generator
import matplotlib.pyplot as plt # this library lets us plot the results
plt.ion() # tell python to use interactive mode for the figure

# generate some random integers
x = np.random.randint(0,20,size=50)

# generate some random real numbers
y = np.random.normal(loc=10.0,scale=5.0,size=50)

# plot the data series
f0 = plt.figure()
p1 = plt.plot(x, label='x')
p2 = plt.plot(y, label='y')
plt.grid()
plt.legend(loc='lower right')
plt.xlabel('Number of the point in the series')
plt.ylabel('Variable')
plt.title('Values of x and y')
plt.show();
````

## 2: Visualising data

Above, we plotted the data, ```x``` and ```y```, as function of their order in the data series. If we had collected data from an instrument, and we wanted to see how it varied over time, we might make a plot of this type.

We might also want to know how likely it is that a variable takes a particular value. In this case, we discard the time information, and look instead at the frequency distribution. Again, we can plot this using python:

````python
f1 = plt.figure()
plt.hist(x,bins=10)
plt.xlabel('x')
plt.ylabel('Frequency')
plt.title('Histogram of x')
plt.show();
````

````python
# enter code in this cell to plot a histogram of y
#bins correspond to the number of boxes, so the number of intervalls on your x-axis
f2 = plt.figure()
plt.hist(y,bins=10)
plt.xlabel('y')
plt.ylabel('Frequency')
plt.title('Histogram of y')
plt.show();
````

````python
#Try increasing and decreasing the number of bins in the two examples above to get an understanding of how this affects the shape of the histogram.
f3 = plt.figure()
plt.hist(y,bins=35)
plt.xlabel('y')
plt.ylabel('Frequency')
plt.title('Histogram of y')
# plot the data series
plt.show();
````
