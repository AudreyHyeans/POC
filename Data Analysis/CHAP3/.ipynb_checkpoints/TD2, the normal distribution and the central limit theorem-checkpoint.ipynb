# Introduction

In this TD, you will look at sea level data measured here in Brest.

A tide gauge is an instrument that measures sea level. They are a simple type of instrument that have existed for a long time, so some of the longest observational time series come from this type of instrument. You may have seen the one at St. Malo already:
![Maregraphe de la Rance](https://upload.wikimedia.org/wikipedia/commons/thumb/4/4c/Tour-du-maregraphe-de-la-Ra.jpg/1460px-Tour-du-maregraphe-de-la-Ra.jpg)
The instruments have changed many times over the years, a modern one might look like this:
![Maregraphe](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f7/Tidal_gauge_dsc06189.jpg/269px-Tidal_gauge_dsc06189.jpg)
Water enters into the bottom of the tube, and, in modern instruments, the sea level is recorded by electronic sensors.

The time series from Brest is one of the longest in the world: measurements have been taken here since 1679 ! 

The data that you will analyse in this TD begin in 1846. 

## 1. Reading in the data

The cell below will read in and prepare the data for you.

````python
# load the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.ion()

# read in the CSV data file from the internet
fname = "http://uhslc.soest.hawaii.edu/woce/d822.dat"
col_list = [i for i in range(1,14)]
df = pd.read_fwf(fname,colspecs=[(0,9),(11,15),(15,19),(20,24),(24,29),\
                                (29,34),(34,39),(39,44),(44,49),(49,54),\
                                 (54,59),(59,64),(64,69),(69,74)])

# clean up and prepare the time data
df.columns = range(df.shape[1])
month = df[2].values//10
df[2] = month
times = pd.date_range(start=str(df[1].values[0])+'-'+str(df[2].values[0]),\
                      end=str(df[1].values[-1])+'-'+str(df[2].values[-1]))
years = np.unique(times.year)
sla_raw = df.values[:,3:]

# create new 1D array containing the sea level as a continuous series
slvl = np.ones(np.shape(times))*9999

for y in years:
    for mon in range(1,13):
        ix = np.where((df[1] == y) & (df[2] == mon))[0]
        i2 = np.where((times.year == y) & (times.month == mon))[0]
        slvl[i2] = np.ravel(sla_raw[ix,:])[:len(i2)]

# set missing data to NaN
slvl[slvl > 9998] = np.nan
````

You should now have 2 variables available to use: ```times```, containing the time information in datetime format, and ```slvl```, containing the sea level. Make a time series plot below to see how the sea level has changed over time since 1846 (there will be some gaps: there is some missing data).

````python
# enter your code here
plt.figure()
plt.title('Sea level from 1847 to 2021')
plt.plot(times, slvl)
plt.xlabel('time in years')
plt.ylabel('sea levels in meters')
plt.show()
````

The missing data in the time series are represented by the value ```NaN```, meaning "Not a Number". Some python functions either cannot cope with ```NaN``` values, and will not work, or will return a value of ```NaN``` when used on an array containing a NaN value. For example, if you use ```np.mean``` on an array containing ```NaN```, it will return ```NaN```, rather than ignoring the missing value.

There are several possible ways to select non-```NaN``` values:

First option: use boolean selection:
- ```x[x==x]``` will select finite values of ```x```
- ```time[x==x]``` will select the values of the ```time``` array for which the corresponding values of ```x``` are finite

Second option: use ```np.isfinite```:
- ```x[np.isfinite(x)]``` will select finite values of ```x```
- ```time[np.isfinite(x)]``` will select the values of the ```time``` array for which the corresponding values of ```x``` are finite

Some functions also have alternative versions that will ignore ```NaN``` values in the input array. For example, ```np.nanmean``` is the equivalent of ```np.mean``` that will return the mean of only non-```NaN``` values.

````python
times = times[np.isfinite(slvl)]
slvl = slvl[np.isfinite(slvl)]
plt.plot(times, slvl)
plt.show()
````
