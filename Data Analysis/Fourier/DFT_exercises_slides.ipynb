```python
#define parameters to compute dfft
N = 2**8

#n has to be n<N/2
n = N/8

#frequency 
Fs = 2

#time T
T = 0.5*N/Fs

#frequency center
fc = np.arange(-Fs/2,Fs/2,1/T)

#time 
t = np.arange(0,N)/Fs

#amplitude
A=1.

#phase
phi = 0.5
print(np.shape(fc))
print(t)
```

```python
#define the function cos
h = A*np.cos(2*np.pi* 1/T *t+phi)
plt.plot(h)
plt.show()
```

```python
#calculate the dfft
H = fft(h)
plt.plot(np.real(H), 
        label = 'real part')
plt.plot(np.imag(H),
        label = 'imaginary part')
plt.legend()
plt.grid()
plt.show()
```

```python
Ng = 2**8
sigma = 1
Fsg = 4
tg = np.arange(-(Ng/2) +1,(Ng/2))*Fsg
```

```python
for sigma in [4,8,10,12]:
    gauss = 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-tg**2/(2*sigma**2))
    plt.plot(tg,gauss)
plt.show()
```

```python
for sigma in [4,8,10,12]:
    gauss = 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-tg**2/(2*sigma**2))
    GAUSS = fft(gauss)
    plt.plot(np.real(GAUSS))
    plt.plot(np.imag(GAUSS))
plt.show()
```

```python
#cumulative phase
sigma = 1
gauss = 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-tg**2/(2*sigma**2))
GAUSS = fft(gauss)
plt.figure()
plt.title('real and imaginary parts')
plt.plot(np.real(GAUSS),label = 'real part')
plt.plot(np.imag(GAUSS),label = 'imaginary part')
plt.xlim(0,50)
plt.legend(loc=2)
plt.show()

```

```python
angle =  np.angle(GAUSS)
phase = np.unwrap(angle)

plt.figure(figsize=(10,5))
plt.title('phase')
plt.plot(phase)

plt.show()
```
