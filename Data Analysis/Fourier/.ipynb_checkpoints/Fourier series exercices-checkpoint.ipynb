# Fourier series exercises 

## Goal : fine the fourier series of the triangular function and saw-toohth function. 

```python
import numpy as np
import matplotlib.pyplot as plt
import signal
```

```python
#period
T = 1
#step delta 
#delta_t = 1/(10*f)
#parameter N
N=10
#time domain where the function is define
t = np.linspace(-1.5*T,1.5*T,100)
#frequence domain where the function is define
f = np.arange(-N,N+1,1)/T
# width 
L = 1*T
M = len(t)

print(np.shape(t), np.shape(f))
def triangular(t):
        y= 1- (2*np.abs(t)) /T
        return(y)

plt.figure()
plt.plot(t,triangular(t))  
plt.xlim(-0.4,0.4)
plt.ylim(0,1)
plt.show();
```

```python
#let's define coefficients Fourier series
t1 = np.linspace(-2*np.pi,2*np.pi,100)
harmonic = 0
for i in range(-N,N):
    harmonic+= 0.5* (np.sinc(i/2)**2) * np.exp( (1j*2*np.pi*i*t)/T) 
#print(c_m)
#im = plt.imread("external-content.duckduckgo.com.jpeg")
#implot = plt.imshow(im)
plt.plot(t,np.real(harmonic), '*', color='black')
plt.plot(t,np.imag(harmonic), '*', color='red')
plt.plot(t,triangular(t))
plt.plot(0.28*np.cos(t1-0.5),0.2*np.sin(t1-0.5)+0.4)
plt.plot(0.1*np.cos(t1-0.5),0.15*np.sin(t1-0.5)+0.4)
plt.scatter (0.,0.4,color='red')
plt.show()
```

# Plot Fourier series of a triangular function 

```python
# for saw-tooth function
from scipy import signal

N = 10
n = np.arange (-N,N,1)
n = n[n!=0]

#I create a meshgird in order to have 2 arrays of same dimensions
tt,nn = np.meshgrid(t,n)

#coefficients
c_n = ((-1.0)**nn) * (1j*T /(2*np.pi*nn))
#Fourier_series
F_series = np.sum(c_n * np.exp(1j * nn * 2* tt*np.pi/T ), axis=0)

#figure
fig, ax = plt.subplots(1,2,figsize = (10,5))
#I plot the saw-tooth function
ax[0].plot(t, signal.sawtooth((t-0.5*T)*2*np.pi/T)*0.5,color='purple')
#plotting the coefs
ax[1].stem(n,np.imag(c_n[:,0]))
#plotting the series
ax[0].plot(t,F_series,label='reconstruction')

ax[0].set_title('Saw-tooth function')
ax[1].set_title('Coefficients c_n')

plt.grid(True, which='both')
plt.legend()
plt.show()
```

## plot for box function

```python
t = np.linspace(-1.5*T,1.5*T,100)

N = 20
n = np.arange (-N/2,N/2,1)
n = n[n!=0]

#I create a meshgird in order to have 2 arrays of same dimensions
tt,nn = np.meshgrid(t,n)

#coefficients
c_n_sf = 0.5*(-1j * np.pi )*nn * (np.sinc(T*nn/2))**2
#Fourier_series
F_series = np.sum(c_n_sf * np.exp(1j * nn * 2* tt*np.pi/T ), axis=0)
#print(F_series)

#figure
fig, ax = plt.subplots(1,2,figsize = (10,5))

#plotting the coefs
ax[1].stem(n,np.imag(c_n_sf[:,0]))
#plotting the series
ax[0].plot(t,np.real(F_series),label='reconstruction',color='black')
ax[0].set_title('Square wave function')
plt.legend()
ax[1].set_title('Coefficients c_n')

plt.grid(True, which='both')

plt.show()
print(c_n_sf)
```
