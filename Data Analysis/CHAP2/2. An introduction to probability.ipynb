## 2.1 Samples and populations

- A population is the collection of all possible measurements that could be made
- A sample is the collection of measurements that we actually make

When we use numpy to generate random data, we are creating a sample. Let's look at last week's example of the uniform distribution again:

````python
import numpy as np # this library contains a random number generator
import matplotlib.pyplot as plt # this library lets us plot the results
plt.ion() # tell python to use interactive mode for the figure

# generate a sample of random integers from the uniform distribution
x = np.random.randint(0,20,size=50)

# plot the histogram
plt.hist(x,bins=10)
plt.xlabel('x')
plt.ylabel('Frequency')
plt.title('Histogram of x: uniform distribution')
plt.show();
````

Now we'll calculate the probability of obtaining a value of ```x``` between 10 and 20.

Remember that the probability is given by the integral of the PDF. So for our exponential distribution:

$$
P(10 \leq x \leq 20) = \int_{10}^{20} p(x) dx = \int_{10}^{20} \frac{1}{\beta}e^{-\frac{x}{\beta}} dx
$$

We can solve this analytically to get:  

$$
P(10 \leq x \leq 20) = \left[-e^{-\frac{x}{\beta}}\right]_{10}^{20} = -e^{-4}+e^{-2}
$$

Calculate the numerical value of this theoretical probability below:

````python
# enter your code to calculate the theoretical value here
P10_20 = np.exp(-2) - np.exp(-4)
print(P10_20)
````

## 2.3 The cumulative distribution function (CDF)

The cumulative distribution function for a given value of $x$ is the integral of the PDF from $-\infty$ up to $x$:
$$F(x) = \int_{-\infty}^{x} p(x) dx$$

In the practice exercise above, you have just made an estimate of $\int_{10}^{20} p(x) dx$ for the exponential function. We can use a similar approach to calculate the CDF.

To calculate the CDF, we will want to keep the lower limit fixed, and then add on more and more increments of $\int_{x1}^{x2} p(x) dx$ as $x2$ varies up to ${+\infty}$. 

Here, you will estimate the CDF from the PDF values and bin values that you obtain from the ```plt.hist``` function, so in reality, $x$ will not vary between $\pm\infty$, but rather over the range of bin values.

Finally, let's use the CDF to generate some random data:  
1. Generate a random number between 0 and 1
2. Use the ```np.interp``` function to interpolate between the two selected indices. The x data will be the CDF values, and the y data will be the bin values.

If you run the cell below several times, you will see several different values in the graph.

When you feel comfortable with how the sampling works, you can try changing the code to use your CDF and bin values that you calculated above. You will then be producing samples from the exponential distribution.

````python
# example CDF and bin data are used here; you can change this example to use your
#  CDF once you feel comfortable with what the code is doing
CDF = [0,0.1,0.4,0.6,0.9,1]
bin_midpoints = np.linspace(0,10,6)

# 1: generate a random number between 0 and 1
nrand = np.random.uniform(low=0,high=1)
print('Sampling at: CDF={:5.3}'.format(nrand))

# 2: interpolate to chosen value
sample_value = np.interp(nrand, CDF, bin_midpoints)
print('Sample value is: {:5.3}'.format(sample_value))

# make a figure to show what happens
plt.figure()
p0=plt.plot(bin_midpoints,CDF,'.-',label='CDF') # our example CDF function
p1=plt.plot([0,sample_value],[nrand,nrand],'r--',label='Sampled point')
p2=plt.plot([sample_value,sample_value],[0,nrand],'r--')
plt.legend();
plt.xlabel('x')
plt.ylabel('Cumulative probability')
plt.xlim(np.min(bin_midpoints),np.max(bin_midpoints)) # set the limits on the x-axis
plt.ylim(0,1)
plt.show(); # set the limits on the x-axis
````
